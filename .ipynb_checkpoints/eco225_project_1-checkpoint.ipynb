**Introduction**

Housing price appreciation is influenced by factors like school quality, state education funding, and economic indicators. This project investigates how private and public school student-to-teacher ratios, enrollment rates, state education funding, income, and consumption affect housing price appreciation across U.S. states from 2009 to 2017. The key question is: Which factors—school quality, state education funding, or economic indicators—are most predictive of housing price appreciation?

Data for this analysis comes from reliable sources: Kaggle's U.S. Schools Dataset (Andrew, 2021) for school data, and U.S. Real Estate Dataset (Sakib, 2021) for housing prices, NCES (n.d.) for state education funding, and the U.S. Bureau of Economic Analysis (BEA, n.d.) for economic indicators. These datasets cover U.S. states from 2009 to 2017, a period of economic recovery and housing market shifts after the 2008 financial crisis.

Building on Machin’s (2008) research, which linked school quality to housing prices, this study expands the scope by including economic variables like income and consumption. This aims to provide a broader understanding of the factors driving housing price appreciation, offering insights for real estate professionals, and educators.


**Explanation of variables**

The student-to-teacher ratio reflects the level of individualized attention students receive, which is often associated with better educational outcomes. A lower ratio suggests smaller class sizes, which can improve student engagement and academic performance. Private schools often have lower student-to-teacher ratios, which is a factor many parents consider when choosing where to live. Public school ratios will usually vary by state and district, and areas with lower ratios may attract more families, increasing housing demand and prices.

Enrollment rates indicate demand for each type of school, which may signal public preference and perceived quality. High-quality schools are attractive to homebuyers, often driving up local housing prices. A higher private school enrollment rate may indicate dissatisfaction with public schools or higher-income households willing to pay for private education, both of which can influence housing demand. Conversely, a strong public school system with high enrollment may make certain neighborhoods more desirable, leading to higher home prices. These enrollment metrics help capture school preference trends, which can be a significant driver of local real estate values.

State revenue per student represents how much financial support each student receives, which impacts school resources, teacher salaries, infrastructure, and overall educational quality. A higher state revenue per student implies stronger state investment in education, which can make school districts within a particular state more attractive on average to families and drive up housing demand in those areas.

Total revenue represents the sum of federal, state, and local funds allocated to schools within a state. Its a  broader measure capturing the overall financial support schools receive from all government levels. Higher total revenue is associated with greater resources available for education, including better facilities, technology, and programs. Since homebuyers prioritize access to quality education, well-funded school districts often correlate with higher property values. 

Total expenditures is the total spending made by schools within a state. This includes teacher salaries, administrative costs, infrastructure, and educational programs. Higher expenditures is typically correlated with better-funded schools, which may lead to improved academic performance and school quality. Therefore, states with greater districts with higher educational spending may experience greater housing price appreciation since homebuyers often consider the strength of local schools when choosing where to live.

Economic indicators include income and consumption, which measure the financial health of a region. Income represents the real personal per capita income. Areas with higher incomes indicate greater purchasing power often driving up house prices resulting in stronger housing demand. Additionally, higher income can correlate with better-funded school districts through local taxes, furthering the connection between economic prosperity and real estate values.  

Consumption represents real personal per capital consumption, reflecting the spending patterns and economic activity of consumers by state, signaling overall prosperity. Higher consumption levels suggest a strong economy where people are willing to spend more on goods, services, and housing. When residents have high disposable income and actively participate in the economy, housing markets tend to be more competitive, leading to price increases.

```python
# Import required libraries for analysis.
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import linregress
import seaborn as sns

# READ & CLEAN DATASETS

# Read private & public school datasets
priv = pd.read_csv('Private_Schools.csv')
pub = pd.read_csv('Public_Schools.csv')

# Read real estate dataset
re = pd.read_csv('realtor-data.zip.csv')

# Read state education funding dataset
df = pd.read_csv('/Users/capi/Downloads/School Expenditure, Salary, & Revenue.csv')

# Read income dataset
df_rpc = pd.read_csv('/Users/capi/Downloads/RPC Income.csv', skiprows=3)

# Read consumption dataset
df_rpcc = pd.read_csv('/Users/capi/Downloads/RPC Consumption.csv', skiprows=3)


# Clean private & public school datasets

# Match column names between datasets
priv.rename(columns = {'SOURCEDATE':'DATE'}, inplace = True)
pub.rename(columns = {'SOURCEDATE':'DATE'}, inplace = True)

priv.dropna(subset = ['STATE', 'CITY', 'DATE', 'ZIP', 'ENROLLMENT', 'FT_TEACHER', 'POPULATION'], inplace = True)
pub.dropna(subset = ['STATE', 'CITY', 'DATE', 'ZIP', 'ENROLLMENT', 'FT_TEACHER', 'POPULATION'], inplace = True)

# Format sourcedate column in school datasets to remove unwanted characters (00:00:00)
priv['DATE'] = pd.to_datetime(priv['DATE']).dt.strftime('%Y/%m/%d')
pub['DATE'] = pd.to_datetime(pub['DATE']).dt.strftime('%Y/%m/%d')

state_mapping = {
    'AL': 'Alabama', 'AK': 'Alaska', 'AZ': 'Arizona', 'AR': 'Arkansas', 'CA': 'California',
    'CO': 'Colorado', 'CT': 'Connecticut', 'DE': 'Delaware', 'FL': 'Florida', 'GA': 'Georgia',
    'HI': 'Hawaii', 'ID': 'Idaho', 'IL': 'Illinois', 'IN': 'Indiana', 'IA': 'Iowa',
    'KS': 'Kansas', 'KY': 'Kentucky', 'LA': 'Louisiana', 'ME': 'Maine', 'MD': 'Maryland',
    'MA': 'Massachusetts', 'MI': 'Michigan', 'MN': 'Minnesota', 'MS': 'Mississippi', 'MO': 'Missouri',
    'MT': 'Montana', 'NE': 'Nebraska', 'NV': 'Nevada', 'NH': 'New Hampshire', 'NJ': 'New Jersey',
    'NM': 'New Mexico', 'NY': 'New York', 'NC': 'North Carolina', 'ND': 'North Dakota', 'OH': 'Ohio',
    'OK': 'Oklahoma', 'OR': 'Oregon', 'PA': 'Pennsylvania', 'RI': 'Rhode Island', 'SC': 'South Carolina',
    'SD': 'South Dakota', 'TN': 'Tennessee', 'TX': 'Texas', 'UT': 'Utah', 'VT': 'Vermont',
    'VA': 'Virginia', 'WA': 'Washington', 'WV': 'West Virginia', 'WI': 'Wisconsin', 'WY': 'Wyoming',
    
    # U.S. Territories
    'PR': 'Puerto Rico', 'VI': 'U.S. Virgin Islands', 'GU': 'Guam',
    'AS': 'American Samoa', 'MP': 'Northern Mariana Islands'
}

# Convert STATE abbreviations to full names in the re dataset
priv['STATE'] = priv['STATE'].map(state_mapping)
pub['STATE'] = pub['STATE'].map(state_mapping)


# Clean real estate dataset

# Match column names with school datasets
re.rename(columns = {'status':'STATUS', 'price':'PRICE', 'city':'CITY', 'state':'STATE', 'zip_code':'ZIP', 'prev_sold_date':'DATE'}, inplace = True)

# Drop rows with missing values from CITY, PRICE, DATE, and ZIP since they will be used in analysis
re.dropna(subset = ['STATE', 'CITY', 'PRICE', 'DATE', 'ZIP', 'STATUS'], inplace = True)

# Drop rows with zero
re = re[re['PRICE'] > 0]

# Capitalize all letters for cities in CITY column real estate dataset
re['CITY'] = re['CITY'].str.upper()

# Format date column for real estate dataset to match date format of school datasets
re['DATE'] = pd.to_datetime(re['DATE'], errors = 'coerce').dt.strftime('%Y/%m/%d')


# Clean income dataset

# Rename columns for clarity
df_rpc = df_rpc.rename(columns={"GeoFips": "GEOFIPS", "GeoName": "STATE"})

# Remove the last 3 rows from the dataset
df_rpc = df_rpc.iloc[:-3]

# Convert from wide format to long format, keeping only years 2009-2017
rpc_inc = df_rpc.melt(id_vars=["GEOFIPS", "STATE"], var_name="DATE", value_name="INCOME")

# Convert DATE column to integer and filter for years 2009-2017
rpc_inc["DATE"] = rpc_inc["DATE"].astype(int)
rpc_inc = rpc_inc[rpc_inc["DATE"].between(2009, 2017)]

# Drop rows with zero
rpc_inc = rpc_inc[rpc_inc["INCOME"] > 0]


# Clean consumption dataset

# Rename columns for clarity
df_rpcc = df_rpcc.rename(columns={"GeoFips": "GEOFIPS", "GeoName": "STATE"})

# Remove the last 3 rows from the dataset
df_rpcc = df_rpcc.iloc[:-3]

# Convert from wide format to long format, keeping only years 2009-2017
rpc_con = df_rpcc.melt(id_vars=["GEOFIPS", "STATE"], var_name="DATE", value_name="CONSUMPTION")

# Convert DATE column to integer and filter for years 2009-2017
rpc_con["DATE"] = rpc_con["DATE"].astype(int)
rpc_con = rpc_con[rpc_con["DATE"].between(2009, 2017)]

rpc_con = rpc_con[rpc_con["CONSUMPTION"] > 0]


# Clean state education funding dataset

columns_mapping = {
    "State": "STATE",
    "Mailing ZIP": "ZIP",
    "Total Salary Expenditures (E11+TE21 +E3A11+E3B11) [State Finance]": "TOTAL SALARY EXPENDITURE",
    "Total Revenue (TOTALREV) [District Finance]": "TOTAL REVENUE",
    "Total Expenditures (TOTALEXP) [District Finance]": "TOTAL EXPENDITURES",
    "State Revenues (R3) per Pupil (MEMBR) [State Finance]": "STATE_REV_PER_STUDENT",
    "Instruction Expenditures - Salaries (E11) [State Finance]": "INSTRUCTION_EXPENDITURE_SALARY",
    "Instruction Expenditures - Tuition (E14) [State Finance]": "INSTRUCTION_EXPENDITURE_TUITION"
}

df = df.rename(columns=columns_mapping)

salary_columns = [col for col in df.columns if "Total Salary Expenditures" in col]
revenue_columns = [col for col in df.columns if "Total Revenue" in col]
expenditure_columns = [col for col in df.columns if "Total Expenditures" in col]
state_rev_columns = [col for col in df.columns if "State Revenues (R3) per Pupil" in col]
instruction_salary_columns = [col for col in df.columns if "Instruction Expenditures - Salaries" in col]
instruction_tuition_columns = [col for col in df.columns if "Instruction Expenditures - Tuition" in col]

# Melting the dataset into long format
df_long = df.melt(id_vars=["STATE", "ZIP"], 
                  value_vars=salary_columns + revenue_columns + expenditure_columns + state_rev_columns + instruction_salary_columns + instruction_tuition_columns,
                  var_name="CATEGORY_YEAR",
                  value_name="VALUE")

df_long["DATE"] = df_long["CATEGORY_YEAR"].str.extract(r'(\d{4})')
df_long["CATEGORY"] = df_long["CATEGORY_YEAR"].str.extract(r'([A-Za-z\s\(\)-]+)')
df_final = df_long.pivot_table(index=["STATE", "ZIP", "DATE"], columns="CATEGORY", values="VALUE", aggfunc="sum").reset_index()

# Renaming columns using the mapping
df_final.columns.name = None  # Remove the multi-index column name
df_final = df_final.rename(columns={
    "TOTAL SALARY EXPENDITURES": "TOTAL SALARY EXPENDITURE",
    "TOTAL REVENUE": "TOTAL REVENUE",
    "TOTAL EXPENDITURES": "TOTAL EXPENDITURES",
    "STATE REVENUES (R3) PER PUPIL": "STATE_REV_PER_STUDENT",
    "INSTRUCTION EXPENDITURES - SALARIES": "INSTRUCTION_EXPENDITURE_SALARY",
    "INSTRUCTION EXPENDITURES - TUITION": "INSTRUCTION_EXPENDITURE_TUITION"
})

# Filtering only for years 2009 to 2017
df_final = df_final[df_final["DATE"].astype(str).between("2009", "2017")]
df_final.columns = df_final.columns.str.strip()

df_final = df_final.rename(columns={
    "Instruction Expenditures - Salaries (E": "INST_EXP_SAL",
    "Instruction Expenditures - Tuition (E": "INST_EXP_TUIT",
    "State Revenues (R": "STATE_REV_PER_STUDENT",
    "Total Expenditures (TOTALEXP)": "TOTALEXP",
    "Total Revenue (TOTALREV)": "TOTALREV",
    "Total Salary Expenditures (E": "TOTAL_SAL_EXP"
})

df_final["STATE"] = df_final["STATE"].str.title()

# Remove rows with NaN values
df_final_cleaned = df_final.dropna()

# Remove rows containing '–' or '‡' in any column
df_final_cleaned = df_final_cleaned.replace(["–", "‡"], pd.NA).dropna()

# Convert relevant columns to numeric
numeric_columns = ["STATE_REV_PER_STUDENT", "TOTALEXP", "TOTALREV"]
df_final_cleaned[numeric_columns] = df_final_cleaned[numeric_columns].apply(pd.to_numeric, errors="coerce")

# Drop rows with zero
df_final_cleaned = df_final_cleaned[df_final_cleaned['STATE_REV_PER_STUDENT'] > 0]
df_final_cleaned = df_final_cleaned[df_final_cleaned['TOTALEXP'] > 0]
df_final_cleaned = df_final_cleaned[df_final_cleaned['TOTALREV'] > 0]


# Calculate private student-to-teacher ratio, public student-to-teacher ratio, 
# private enrollment rate, and public enrollment rate variables

# Calculate student-to-teacher ratio in private schools for each state
priv['Priv_ST_Ratio'] = priv['ENROLLMENT'] / priv['FT_TEACHER']
priv['Priv_ST_Ratio'].replace([np.inf, -np.inf], np.nan, inplace=True)
priv.dropna(subset=['Priv_ST_Ratio'], inplace=True)
priv = priv[priv['Priv_ST_Ratio'] > 0]
priv['Priv_ST_State'] = priv.groupby('STATE')['Priv_ST_Ratio'].transform('mean')

# Calculate student-to-teacher ratio in public schools for each state
pub['Pub_ST_Ratio'] = pub['ENROLLMENT'] / pub['FT_TEACHER']
pub['Pub_ST_Ratio'].replace([np.inf, -np.inf], np.nan, inplace=True)
pub.dropna(subset=['Pub_ST_Ratio'], inplace=True)
pub = pub[pub['Pub_ST_Ratio'] > 0]
pub['Pub_ST_State'] = pub.groupby('STATE')['Pub_ST_Ratio'].transform('mean')

# Calculate enrollment rate in private schools for each state
priv['Priv_ER'] = priv['ENROLLMENT'] / priv['POPULATION']
priv['Priv_ER'].replace([np.inf, -np.inf], np.nan, inplace=True)
priv.dropna(subset=['Priv_ER'], inplace=True)
priv = priv[priv['Priv_ER'] > 0]
priv['Priv_ER_State'] = priv.groupby('STATE')['Priv_ER'].transform('mean')

# Calculate enrollment rate in private schools for each state
pub['Pub_ER'] = pub['ENROLLMENT'] / pub['POPULATION']
pub['Pub_ER'].replace([np.inf, -np.inf], np.nan, inplace=True)
pub.dropna(subset=['Pub_ER'], inplace=True)
pub = pub[pub['Pub_ER'] > 0]
pub['Pub_ER_State'] = pub.groupby('STATE')['Pub_ER'].transform('mean')


# SUMMARY STATISTICS OF X-VARIABLES & Y-VARIABLE

priv_summary_stats = priv[['Priv_ST_State', 'Priv_ER_State']].describe()

pub_summary_stats = pub[['Pub_ST_State', 'Pub_ER_State']].describe()

rpc_inc_summary_stats = rpc_inc[['INCOME']].describe()

rpc_con_summary_stats = rpc_con[['CONSUMPTION']].describe()

df_final_cleaned_summary_stats = df_final_cleaned[['STATE_REV_PER_STUDENT', 'TOTALEXP', 'TOTALREV']].describe()

re_summary_stats = re[['PRICE']].describe()

# Create a summary table with statistics for variables
summary_table = pd.DataFrame()

# Define datasets and variable groups
datasets = {
    "Private Schools": priv[["Priv_ST_State", "Priv_ER_State"]],
    "Public Schools": pub[["Pub_ST_State", "Pub_ER_State"]],
    "Income": rpc_inc[["INCOME"]],
    "Consumption": rpc_con[["CONSUMPTION"]],
    "Education Funding": df_final_cleaned[["STATE_REV_PER_STUDENT", "TOTALEXP", "TOTALREV"]],
    "Real Estate": re[["PRICE"]]
}

# Generate summary statistics for each dataset
for name, data in datasets.items():
    summary_stats = data.describe().T 
    summary_stats["Dataset"] = name  
    summary_table = pd.concat([summary_table, summary_stats])

# Rename columns
summary_table = summary_table.rename(columns={
    "count": "Observations",
    "mean": "Mean",
    "std": "Std Dev",
    "min": "Min",
    "25%": "Q1 (25%)",
    "50%": "Median",
    "75%": "Q3 (75%)",
    "max": "Max"
})

# Move dataset column to the front
summary_table = summary_table.reset_index().rename(columns={"index": "Variable"})
summary_table = summary_table[["Dataset", "Variable", "Observations", "Mean", "Std Dev", "Min", "Q1 (25%)", "Median", "Q3 (75%)", "Max"]]

# Define a function to format numbers for better readability
def format_large_numbers(value):
    if isinstance(value, (int, float)):
        if abs(value) >= 1e6:
            return f"{value/1e6:.2f}M"  # Format as millions
        elif abs(value) >= 1e3:
            return f"{value/1e3:.2f}K"  # Format as thousands
        else:
            return f"{value:.2f}"  
    return value

# Apply formatting to the summary table
summary_table_formatted = summary_table.copy()
summary_table_formatted.iloc[:, 2:] = summary_table_formatted.iloc[:, 2:].applymap(format_large_numbers)

summary_table_formatted
```

Private schools tend to have small class sizes with 10 students per teacher across all states on average, which is associated with higher quality of education.
```python
# Merge the datasets based on "STATE" column to have both "STATE_REV_PER_STUDENT" and "PRICE" in the same dataframe
merged_data = pd.merge(df_final_cleaned[['STATE', 'STATE_REV_PER_STUDENT']], re[['STATE', 'PRICE']], on='STATE')

# Aggregate by state, calculate the mean for each state
agg_data = merged_data.groupby('STATE').agg({
    'STATE_REV_PER_STUDENT': 'mean',
    'PRICE': 'mean'
}).reset_index()

# Perform linear regression
slope, intercept, r_value, p_value, std_err = linregress(agg_data['STATE_REV_PER_STUDENT'], agg_data['PRICE'])

# Create a scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(agg_data['STATE_REV_PER_STUDENT'], agg_data['PRICE'], alpha=0.7, color='b')

# Add a regression line
plt.plot(agg_data['STATE_REV_PER_STUDENT'], slope * agg_data['STATE_REV_PER_STUDENT'] + intercept, color='r', label=f'Linear regression: y = {slope:.2f}x + {intercept:.2f}')

plt.xlabel('State Revenue per Student ($ Dollars)')
plt.ylabel('Housing Price ($ Millions)')
plt.title('Comparing State Revenue per Student on Housing Price Across all US States from 2009-2017')

# Show the regression line equation in the legend
plt.legend()

plt.grid(True)
plt.show()
```

There is a positive relationship between state revenue per student and housing prices. As the revenue per student increases, housing prices tend to increase as well, suggesting that states with higher revenue per student tend to have higher housing prices. This supports that state investment in education could make areas more desirable, thus leading to price appreciation in housing.

```python
# Group the data by year and calculate the average income and consumption for each year
avg_inc_per_year = rpc_inc.groupby('DATE')['INCOME'].mean()
avg_con_per_year = rpc_con.groupby('DATE')['CONSUMPTION'].mean()

# Create a dataframe for the plot
years = avg_inc_per_year.index
avg_values = {
    'Income': avg_inc_per_year.values,
    'Consumption': avg_con_per_year.values
}

# Plot the bar chart
fig, ax = plt.subplots(figsize=(12, 6))

bar_width = 0.35
index = np.arange(len(years))

bar1 = ax.bar(index - bar_width/2, avg_values['Income'], bar_width, label='Income', color='skyblue')
bar2 = ax.bar(index + bar_width/2, avg_values['Consumption'], bar_width, label='Consumption', color='salmon')

ax.set_xlabel('Year')
ax.set_ylabel('Average Value ($)')
ax.set_title('Comparing State Average Real Personal per Capita Income and Consumption Over Time in the US')
ax.set_xticks(index)
ax.set_xticklabels(years)
ax.legend()

plt.tight_layout()
plt.show()
```

The chart shows that income is consistently higher than consumption in each year, though both increase slightly over time. States with higher incomes and consumption could potentially have higher housing prices, as people with more disposable income tend to be willing to pay more for housing.
